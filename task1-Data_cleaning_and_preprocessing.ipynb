```python
# This line creates a new column "Full Name" by combining "First Name" and "Last Name" in a pandas DataFrame.
df['Full Name'] = df['First Name'] + ' ' + df['Last Name']
```

```python
df
```

```python
# This line removes specific columns from your DataFrame in pandas.
df=df.drop(columns=['First Name','Last Name','Customer Id','Phone 2'])
df
```

```python
# Count missing values in each column of a DataFrame.
df.isna().sum()
```

```python
# This line converts a column into proper date/time format using pandas.
df['Subscription Date'] = pd.to_datetime(df['Subscription Date'])
df
```

```python
# This code is using pandas to convert multiple text columns to lowercase.
cols = ['Company', 'City', 'Country','Full Name']

df[cols] = df[cols].apply(lambda x: x.str.lower())
df
```

```python
# This line cleans a phone number column using pandas by removing all non-digit characters.
df['Phone 1'] = df['Phone 1'].str.replace(r'\D', '', regex=True)
df
```

```python
# Count missing values in each column of a DataFrame.
df.isna().sum()
```
